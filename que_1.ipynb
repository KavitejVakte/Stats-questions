Q1) Write a Python program using Pandas and Matplotlib to analyze student scores, calculate descriptive statistics and mode, and visualize the data using a histogram with mean and median lines.  data = [23, 33, 39, 43, 45, 43, 45, 55, 82, 88]

```python
import numpy as np
from scipy import stats

# Sample dataset (e.g., marks)
data = [23, 33, 39, 43, 45,
        43, 45, 55, 82, 88, 88]

# Mean
mean = np.mean(data)
print("Mean =", mean)

# Median
median = np.median(data)
print("Median =", median)

# Mode
mode_r = stats.mode(data, keepdims=True)
print("Mode =", mode_r.mode[0],
      "(freq =", mode_r.count[0], ")")

# Weighted Mean
weights = [1, 2, 3, 2, 3, 6, 4, 4, 2, 1, 1]
w_mean = np.average(data, weights=weights)

print("W Mean =", round(w_mean, 2))
```
