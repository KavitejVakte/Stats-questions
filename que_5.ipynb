Write a Python program to calculate the Population Variance & Standard Deviation and Sample Variance & Standard Deviation using NumPy.

```python
import numpy as np

# Given marks
marks = np.array([4, 8, 6, 5, 7])

# Population (divide by N) - ddof = 0 (default)
print("Population Variance:", np.var(marks))
print("Population Standard Deviation:", np.std(marks))

# Sample (divide by n-1) - ddof = 1
print("Sample Variance:", np.var(marks, ddof=1))
print("Sample Standard Deviation:", np.std(marks, ddof=1))
```
