Q2)Write a Python program using NumPy and SciPy to calculate the mean, median, mode, and weighted mean of a given dataset and display the results.data = [23, 33, 39, 43, 45, 43, 45, 55, 82, 88]

```python
import numpy as np
from scipy import stats

# Sample data
data = [23, 33, 39, 43, 45,
        43, 45, 55, 82, 88]

# Mean
mean = np.mean(data)
print(f"Mean : {mean:.2f}")

# Median
median = np.median(data)
print(f"Median : {median}")

# Mode
mode = stats.mode(data, keepdims=True)
print(f"Mode : {mode.mode[0]}")

# Weighted Mean
weights = [1, 2, 3, 2, 3, 4, 2, 1, 1, 1]
w_mean = np.average(data, weights=weights)

print(f"W Mean : {w_mean:.2f}")
```
