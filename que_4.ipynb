Write a Python program to detect outliers using the 1.5 × IQR rule for the given dataset.

```python
import numpy as np

# Dataset
data = np.array([35, 41, 44, 45, 47, 48, 49, 50, 52, 53, 55, 58, 62, 71, 92])

# Quartiles
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

# IQR
iqr = q3 - q1

# Fences
low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr

# Outliers
outliers = data[(data < low) | (data > high)]

print("Q1 =", q1)
print("Q3 =", q3)
print("IQR =", iqr)
print("Lower Fence =", low)
print("Upper Fence =", high)
print("Outliers =", outliers)
```
