Question

Write a Python program to calculate the Range and Interquartile Range (IQR) of the following dataset using NumPy:

[3, 5, 7, 8, 12, 13, 14, 18, 21]

Also display:

Range
First Quartile (Q1)
Third Quartile (Q3)
Interquartile Range (IQR)

```python
import numpy as np

scores = np.array([3, 5, 7, 8, 12, 13, 14, 18, 21])

# Range
data_range = scores.max() - scores.min()

# Quartiles and IQR
q1 = np.percentile(scores, 25)
q3 = np.percentile(scores, 75)
iqr = q3 - q1

print(f"Range = {data_range}")
print(f"Q1 = {q1}, Q3 = {q3}, IQR = {iqr}")
```
